```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: "setosa", 1: "versicolor", 2: "virginica"})

df.head()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="sepal length (cm)", y="sepal width (cm)")
plt.title("Scatterplot without categorical separation")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, 
                x="sepal length (cm)", 
                y="sepal width (cm)", 
                hue="species", palette="Set1")
plt.title("Scatterplot with categorical separation using color")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df,
                x="petal length (cm)",
                y="petal width (cm)",
                hue="species",
                style="species",   # different markers
                s=80)              # size
plt.title("Scatterplot with categorical ordering (markers + colors)")
plt.show()
```

```python
g = sns.FacetGrid(df, col="species", hue="species", height=4, palette="Set1")
g.map_dataframe(sns.scatterplot, x="petal length (cm)", y="petal width (cm)")
g.add_legend()
plt.subplots_adjust(top=0.8)
g.fig.suptitle("Scatterplots aligned by category (Facets)")
plt.show()
```
