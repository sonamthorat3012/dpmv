```python
df.plot()
```

```python
df_numeric = df.select_dtypes(include='number')
print("Numeric columns:", df_numeric.columns.tolist())
df_numeric.plot(
    kind='box',
    subplots=True,
    layout=(1, len(df_numeric.columns)),  # all in one row
    figsize=(15, 5)
)
plt.tight_layout()
plt.show()
```

```python
colors = {
    'boxes': 'skyblue',
    'whiskers': 'gray',
    'medians': 'red',
    'caps': 'black'
}

df_numeric.plot(
    kind='box', 
    color=colors,   # apply colors
    patch_artist=True,  # fill the box with color
    subplots=True, 
    layout=(1, len(df_numeric.columns)),
    figsize=(15, 5)
)
plt.tight_layout()
plt.show()
```
