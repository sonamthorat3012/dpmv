```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ------------------------------
# 1. LOAD DATASET
# ------------------------------
df = pd.read_csv("diabetes.csv")

# Only categorical variable in dataset = Outcome
# Create another categorical variable (BMI Category)
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["BMI_Category"] = df["BMI"].apply(bmi_category)
```

```python
print("\n====== FREQUENCY OF OUTCOME ======")
print(df["Outcome"].value_counts())

print("\n====== FREQUENCY OF BMI CATEGORY ======")
print(df["BMI_Category"].value_counts())

# --- BAR PLOTS (frequency comparison) ---

plt.figure(figsize=(7,5))
sns.countplot(x=df["Outcome"])
plt.title("Frequency Distribution of Outcome")
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(7,5))
sns.countplot(x=df["BMI_Category"])
plt.title("Frequency Distribution of BMI Categories")
plt.xlabel("BMI Category")
plt.ylabel("Count")
plt.show()
```

```python
#B) PROPORTION OF EACH CATEGORY (PIE CHARTS)
# ---------------------------------------------------

plt.figure(figsize=(6,6))
df["Outcome"].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Proportion of Outcome Categories")
plt.ylabel("")
plt.show()

plt.figure(figsize=(6,6))
df["BMI_Category"].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Proportion of BMI Categories")
plt.ylabel("")
plt.show()
```

```python
#C) CATEGORY DISTRIBUTION SPLIT BY ANOTHER CATEGORY
# ---------------------------------------------------

# Example: BMI category split by Outcome
plt.figure(figsize=(8,6))
sns.countplot(x="BMI_Category", hue="Outcome", data=df)
plt.title("BMI Category Distribution Split by Outcome")
plt.xlabel("BMI Category")
plt.ylabel("Count")
plt.legend(title="Outcome")
plt.show()

# Example 2: Outcome split by BMI category
plt.figure(figsize=(8,6))
sns.countplot(x="Outcome", hue="BMI_Category", data=df)
plt.title("Outcome Split by BMI Category")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.legend(title="BMI Category")
plt.show()
```
