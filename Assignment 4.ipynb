```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
# 1. LOAD DATASET

df = pd.read_csv("diabetes.csv")

# Choose any two variables for demonstration
x = df["Glucose"]
y = df["BMI"]
```

```python
# 2. SCATTER PLOT

plt.figure(figsize=(7,5))
plt.scatter(x, y)
plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.title("Scatter Plot: Glucose vs BMI")
plt.show()
```

```python
# 3. HISTOGRAM FOR NORMALITY CHECK

plt.figure(figsize=(7,5))
plt.hist(x, bins=20, edgecolor='black')
plt.title("Histogram of Glucose")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. QQ-PLOT (Graph method for normality)

plt.figure(figsize=(7,5))
stats.probplot(x, dist="norm", plot=plt)
plt.title("QQ Plot of Glucose")
plt.show()
```

```python
# 5. BOX PLOT (Symmetry check)

plt.figure(figsize=(7,5))
sns.boxplot(x=x)
plt.title("Boxplot of Glucose")
plt.show()
```

```python
# 6. STATISTICAL NORMALITY TEST
#    (ShapiroWilk Test)
# -----------------------------
shapiro_stat, shapiro_p = stats.shapiro(x)
print("Shapiro-Wilk Test for Glucose")
print("Statistic:", shapiro_stat)
print("p-value :", shapiro_p)

if shapiro_p > 0.05:
    print("Conclusion: Data appears to be NORMAL (fail to reject H0)")
else:
    print("Conclusion: Data is NOT normal (reject H0)")
```

```python
# 7. SYMMETRY TEST (Skewness)
# -----------------------------
skew_x = stats.skew(x)
print("\nSkewness for Glucose:", skew_x)

if skew_x > 0:
    print("Interpretation: Positively skewed (Right skewed)")
elif skew_x < 0:
    print("Interpretation: Negatively skewed (Left skewed)")
else:
    print("Interpretation: Perfectly symmetric")
```
